```python
# Question: Create a scatter plot using random values/data.
import numpy as np
import matplotlib.pyplot as plt
#Answer
x=np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
plt.scatter (x,y, c=colors, cmap='viridis')
plt.colorbar(label = 'Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.show()
```

```python
# Question: Create a bar plot on the basis of given categories and values
import numpy as np
import matplotlib.pyplot as plt
#Answer
categories = ['A','B', 'C']
values = [10, 25, 32]
plt.bar (categories, values)
plt.xlabel('Catgories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

```python
# Question: Create a Line plot for the function y = sin(x) and y= cos(x) with Les

#Answer
import numpy as np
import matplotlib.pyplot as plt
# Generate data
x = np. linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np. cos(x)
# Create the plot
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with legends')
plt.show()
```
