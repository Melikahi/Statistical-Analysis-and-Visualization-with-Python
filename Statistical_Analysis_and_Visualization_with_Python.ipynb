```python
#Data
data = np.random.rand(1000)
# Generate data for 1000 samples
import numpy as np
import pandas as pd

np.random.seed(0)

n_samples = 1000

age = np.random.normal(35, 10, n_samples)

height = np.random.normal(170, 15, n_samples)

weight = np.random.normal(70, 10, n_samples)

gender = np.random.choice(['Male', 'Female'], n_samples)

income = np.random.normal(50000, 15000, n_samples)

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Height': height,
    'Weight': weight,
    'Gender': gender,
    'Income': income
})

print(df.head())

```

3. Calculate the mode for Gender

```python

statistics = { 'Age': {'Mean': np.mean(df['Age']),'Median': np.median(df['Age']),'Standard Deviation': np.std(df['Age']),'Variance': np.var(df['Age'])},
    'Height': {'Mean': np.mean(df['Height']),'Median': np.median(df['Height']),'Standard Deviation': np.std(df['Height']),'Variance': np.var(df['Height'])},
    'Weight': {'Mean': np.mean(df['Weight']),'Median': np.median(df['Weight']),'Standard Deviation': np.std(df['Weight']),'Variance': np.var(df['Weight']) },
    'Income': {'Mean': np.mean(df['Income']),'Median': np.median(df['Income']),'Standard Deviation': np.std(df['Income']),'Variance': np.var(df['Income']) }
}

# Assuming df['Gender'] contains non-numeric data
df['Gender'] = df['Gender'].astype(str)

# Using np.unique to get the mode
unique_values, counts = np.unique(df['Gender'], return_counts=True)
mode_gender = unique_values[np.argmax(counts)]


# results
for key, value in statistics.items():
    print(f'{key}:')
    for stat, val in value.items():
        print(f'  {stat}: {val}')
print(f'Gender Mode: {mode_gender}')
```

```python
import seaborn as sns
# Set up the figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 10))

# Histograms
sns.histplot(df['Age'], kde=False, ax=axes[0, 0], color='pink', bins=20)
sns.histplot(df['Height'], kde=False, ax=axes[0, 1], color='blue', bins=20)
sns.histplot(df['Weight'], kde=False, ax=axes[0, 2], color='yellow', bins=20)
sns.histplot(df['Income'], kde=False, ax=axes[0, 3], color='green', bins=20)

# KDE Plots
sns.kdeplot(df['Age'], ax=axes[1, 0], color='blue', fill=True)
sns.kdeplot(df['Height'], ax=axes[1, 1], color='green', fill=True)
sns.kdeplot(df['Weight'], ax=axes[1, 2], color='orange', fill=True)
sns.kdeplot(df['Income'], ax=axes[1, 3], color='red', fill=True)


# Set titles
axes[0, 0].set_title('Age Histogram')
axes[0, 1].set_title('Height Histogram')
axes[0, 2].set_title('Weight Histogram')
axes[0, 3].set_title('Income Histogram')
axes[1, 0].set_title('Age KDE Plot')
axes[1, 1].set_title('Height KDE Plot')
axes[1, 2].set_title('Weight KDE Plot')
axes[1, 3].set_title('Income KDE Plot')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Set up the figure with subplots
fig, axes = plt.subplots(1, 4, figsize=(18, 6))

# Boxplots
sns.boxplot(y=df['Age'], ax=axes[0], color='pink')
sns.boxplot(y=df['Height'], ax=axes[1], color='blue')
sns.boxplot(y=df['Weight'], ax=axes[2], color='yellow')
sns.boxplot(y=df['Income'], ax=axes[3], color='green')

# Set titles
axes[0].set_title('Age Boxplot')
axes[1].set_title('Height Boxplot')
axes[2].set_title('Weight Boxplot')
axes[3].set_title('Income Boxplot')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame with columns Age, Height, Weight, and Income
correlation_matrix = df[['Age', 'Height', 'Weight', 'Income']].corr()

# Extracting the Pearson correlation coefficients
pearson_corr = correlation_matrix.loc['Income', ['Age', 'Height', 'Weight']]

print("Pearson Correlation Coefficients:")
print(pearson_corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from scipy import stats

# Assuming df for 'Income' and 'Gender' columns
income_male = df[df['Gender'] == 'Male']['Income']
income_female = df[df['Gender'] == 'Female']['Income']

# t-test
t_statistic, p_value = stats.ttest_ind(income_male, income_female)

print(f"T-Statistic: {t_statistic}, P-Value: {p_value}")
if p_value < 0.05:
    print("There is a significant difference in Income between Male and Female.")
else:
    print("There is no significant difference in Income between Male and Female.")
```
